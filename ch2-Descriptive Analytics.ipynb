```python
soldprice_by_age = ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age)
```

```python
soldprice_by_age_role =  ipl_auction_df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age_role)
```

## Joining Data frames


```python
sold_price_comparision = soldprice_by_age_role.merge(soldprice_by_age,on= 'AGE', how= 'outer')
```

```python
sold_price_comparision
```

## |   Re-Naming Columns

```python
sold_price_comparision.rename(columns ={'SOLD PRICE_x':'SOLD_PRICE_AGE_ROLE','SOLD PRICE_y':'SOLD_PRICE_AGE'},inplace= True)
```

```python
print(sold_price_comparision)
```

## Applying operations on multiple Columns

```python
sold_price_comparision['change'] = sold_price_comparision.apply(lambda rec:(rec.SOLD_PRICE_AGE_ROLE-rec.SOLD_PRICE_AGE)/rec.SOLD_PRICE_AGE,axis = 1)
```

```python
sold_price_comparision
```
